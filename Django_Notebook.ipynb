## 7. Building A RESTful API

### 7.1 What Are RESTful APIs
* API: Application Programming Interface
* Building an interface so that the client can get data from and send data to the web server
* RESTful: REpresentaional State Transfer

* RESTful APIs are built on three concepts:
    1. __Resources:__ It is like an object in the application that are available on the web and accessible through a URLs
    2. __Representation:__ After accessing a resource, it is returned in a certain format like HTML, XML, or JSON which is deferent from the the data representation on a server. On the server, data is stored as an instance of a python class.
        * JSON: JavaScript Object Notation
    3. __HTTP Response:__ Define methods that allows different type of interactions with the resources.
        * GET: to get a resource from the server to the client
        * POST: to create a resource in the server
        * PUT: to update a resource
        * PATCH: to partial update a resource
        * DELETE: to remove a resource from the server
* While data models can change, API models should be consistent

### 7.2 Designing an API
1. Figure out the different operations of the app
2. For each operation figuring out the:
    * HTTP method type
    * url endpoints
    * request type
    * response type

# 9. Securing APIs

#### 9.1 Authentications Process
1. Client sends a POST request to register
2. Server process the POST request and creates a new user
3. Client sends a request to login using credential info
4. Server process the credential info then:
    1. If wrong credentials are given -> return an error
    2. If credentials were OK -> generate and send a Token

* Token is a protected key with access level permissions for a user
* Each time the user try to access a resource, this token is evaluated

* __DRF Token__ vs __Jason Web Token__
    * DRF Token uses a database table to store token and enquire the database for authentication each time
    * JWT tokens are stored in cookies in the browser and uses signatures for validations

* JWT token has three parts:
    1. Header
    2. Payload -> Contains the user ID
    3. Verify Signature -> Generated using Header and Payload

* JWT tokens are generated in the server and cannot be hacked unless accessing the server
* JWT generate two tokens:
    1. access: used for verifying the user and its authentication and permissions
    2. refresh: used for verifying the access token validity

#### 9.2 Adding the Authentication Endpoint
* Django does not include Auth system for API layer, so use:
